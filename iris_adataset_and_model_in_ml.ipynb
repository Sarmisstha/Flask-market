```python
import seaborn as sns
iris=sns.load_dataset('iris')
iris.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
model.fit(X,Y)
```

```python
predictions=model.predict(X)
```
